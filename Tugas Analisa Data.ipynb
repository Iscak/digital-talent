# 1. Deskripsi Data
Data ini berisi mengenai nilai ekspor dan impor negara Indonesia.

Penjelasan mengenai variabel pada data:
<br>a. tahun: Tahun
<br> -> data diambil dari tahun 2005 sampai tahun 2013.
<br>b. bulan: Bulan
<br> -> dari bulan Januari sampai Desember.
<br>c. Benua: Benua
<br> -> benua Amerika, Eropa, Asia dan Timur Tengah, Australia, dan Afrika.
<br>d. region: Region
<br> -> region Amerika Utara, Uni Eropa, ASEAN, Rusia, dll.
<br>e. nama_negara_tujuan: Negara tujuan
<br> -> Negara Amerika Serikat, Inggris, Vietnam, New Zealand, dll.
<br>f. nama_negara_asal: Negara asal
<br> -> Negara Amerika Serikat, Inggris, Vietnam, New Zealand, dll.
<br>g. nilai_ekspor : Nilai ekspor
<br>h. nilai_impor : Nilai impor



```python
# import file
import pandas as pd

data_ekspor = pd.read_csv('nilai-ekspor-negara-tujuan-2005-2013.csv')
data_impor = pd.read_csv('nilai-impor-negara-asal-2005-2013.csv')
```

# 2. Statistik Deskriptif

```python
data_ekspor.tahun.count()
```

```python
data_ekspor.nilai_ekspor.describe()
```

```python

s = data_ekspor.nilai_ekspor.sum()
mi = data_ekspor.nilai_ekspor.min()
ma = data_ekspor.nilai_ekspor.max()
me = data_ekspor.nilai_ekspor.mean()
md = data_ekspor.nilai_ekspor.median()


print('==================================================')
print('Jumlah nilai ekspor adalah',s)
print('==================================================')
print('Nilai minimum',mi)
print('==================================================')
print('Nilai maksimum',ma)
print('==================================================')
print('Rata-rata dari nilai ekspor adalah',me)
print('==================================================')
print('Median dari nilai ekspor adalah',md)
```

```python
s = data_impor.nilai_impor.sum()
mi = data_impor.nilai_impor.min()
ma = data_impor.nilai_impor.max()
me = data_impor.nilai_impor.mean()
md = data_impor.nilai_impor.median()


print('==================================================')
print('Jumlah nilai impor adalah',s)
print('==================================================')
print('Nilai minimum',mi)
print('==================================================')
print('Nilai maksimum',ma)
print('==================================================')
print('Rata-rata dari nilai impor adalah',me)
print('==================================================')
print('Median dari nilai impor adalah',md)
```

# 3. Visualisasi Sederhana

```python
ax = data_ekspor.groupby('tahun').mean()['nilai_ekspor']
ax = ax.plot(kind='line', title ="Rata-Rata Nilai Ekspor Berdasarkan Tahun", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai Ekspor", fontsize=12)
```

```python
ax = data_ekspor.groupby('tahun').mean()['nilai_ekspor']
ax = ax.plot(kind='bar', title ="Rata-Rata Nilai Ekspor Berdasarkan Tahun", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai Ekspor", fontsize=12)
```

```python
ax = data_impor.groupby('tahun').mean()['nilai_impor']
ax = ax.plot(kind='line', title ="Rata-Rata Nilai Impor Berdasarkan Tahun", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai Impor", fontsize=12)
```

```python
ax = data_impor.groupby('tahun').mean()['nilai_impor']
ax = ax.plot(kind='bar', title ="Rata-Rata Nilai Impor Berdasarkan Tahun", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("tahun", fontsize=12)
ax.set_ylabel("nilai_impor", fontsize=12)
```
