```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# One to many relationship

```python
# 1. Report the account representative for each customer.
run("""
SELECT 
    c.customerName,  
    CONCAT(c.contactFirstName,' ',c.contactLastName) AS customerContact,
    CONCAT(e.firstName,' ',e.lastname) AS Sales_Rep
    
FROM
    customers c
        INNER JOIN
    employees e ON c.salesRepEmployeeNumber = e.employeeNumber
""")
```

```python
# 2. Report total payments for Atelier graphique.
run("""
SELECT 
    c.customerName, Sum(amount) AS totalPayment
FROM customers c 
    INNER JOIN 
    payments p ON c.customerNumber=p.customerNumber
WHERE 
    c.customerName LIKE 'Atelier%'
GROUP BY c.customerName
""")
```

```python
# 3. Report the total payments by date
run("""
SELECT 
    paymentDate, Sum(amount) AS totalPayment
FROM payments 
GROUP BY paymentDate
""")
```

```python

# 4. Report the products that have not been sold.
run("""
SELECT 
    p.productName,p.productCode, o.quantityOrdered AS jumlah
FROM products p 
    LEFT JOIN 
    orderdetails o ON o.productCode=p.productCode
WHERE 
    o.quantityOrdered is NULL
""")
```

```python

# 5. List the amount paid by each customer.
run("""
SELECT 
    c.customerName,
    c.customerNumber,
    sum(amount) as Totalbayar
FROM 
    customers c
    INNER JOIN 
    payments p ON c.customerNumber = p.customerNumber
GROUP BY customerNumber
""")
```

```python

# 6. How many orders have been placed by Herkku Gifts?
run("""
SELECT 
    c.customerName, 
    COUNT(o.orderNumber)AS jumlah
FROM customers c 
    INNER JOIN 
    orders o ON c.customerNumber= o.customerNumber
WHERE 
    c.customerName LIKE 'Herkku Gifts'
GROUP BY c.customerName
""")
```

```python

# 7.Who are the employees in Boston?
run("""
SELECT 
    e.employeeNumber,
    CONCAT(e.firstName,' ',e.lastName) AS employee_name,
    o.city
FROM employees e 
    INNER JOIN 
    offices o ON e.officeCode= o.officeCode
WHERE 
    o.city LIKE 'Boston'
""")
```

```python

    
# 8. Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first.
run("""
SELECT  
    c.customerName, 
    sum(p.amount) as payments
FROM 
    customers c 
    INNER JOIN 
    payments p ON c.customerNumber = p.customerNumber
GROUP BY c.customerName
HAVING
    payments > 100000
ORDER BY payments DESC
""")
```

```python
# 9. List the value of 'On Hold' orders.
run("""
SELECT 
    o.orderNumber, 
    o.status,
    SUM(d.quantityOrdered*d.priceEach) AS Total
FROM 
    orders o 
    INNER JOIN 
    orderdetails d USING(orderNumber)
WHERE o.status LIKE "On Hold"
GROUP BY o.orderNumber
""")
```

```python
# 10. Report the number of orders 'On Hold' for each customer.
run("""
SELECT 
    c.customerName, 
    COUNT(o.orderNumber) AS transaction
FROM 
    orders o 
    INNER JOIN 
    customers c USING (customerNumber)
WHERE o.status LIKE 'On Hold'
GROUP BY c.customerName
""")
```

# Many to many relationship

```python
# 1. List products sold by order date
run("""
SELECT 
    o.orderDate,
    p.productName
FROM 
    orderdetails d
    INNER JOIN
    orders o ON o.orderNumber = d.orderNumber
    INNER JOIN
    products p ON d.productCode = p.productCode
""")
```

```python
# 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
run("""
SELECT 
    p.productName, 
    o.orderDate
FROM 
    orderdetails d
    INNER JOIN 
    products p ON d.productCode = p.productCode
    INNER JOIN
    orders o ON o.orderNumber = d.orderNumber
WHERE 
     p.productName = '1940 Ford Pickup Truck'
ORDER BY 
    o.orderDate DESC
""")
```

```python
# 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
run("""
SELECT 
    c.customerName, 
    o.orderNumber, 
    SUM(d.quantityOrdered * d.priceEach) AS total
FROM 
    orders o
    INNER JOIN
    customers c ON o.customerNumber = c.customerNumber
    INNER JOIN 
    orderdetails d ON o.orderNumber = d.orderNumber   
GROUP BY 
    o.orderNumber
HAVING 
    total > '25000'
ORDER BY total
""")
```

```python
# 4. Are there any products that appear on all orders?
run("SELECT COUNT(DISTINCT orderNumber) FROM orderdetails")
```

```python
run("""
SELECT 
    d.productCode,p.productName
FROM 
    (SELECT COUNT(orderNumber) AS jumlah FROM orders) AS o
    INNER JOIN
    (SELECT productCode, COUNT(orderNumber) AS jumlah FROM orderdetails GROUP BY productCode)  AS d  
    INNER JOIN 
    (SELECT productName FROM products) AS p
WHERE 
    o.jumlah = d.jumlah

""")
```

```python
# 5. List the names of products sold at less than 80% of the MSRP.
run("""

SELECT  
    p.productName, 
    d.PriceEach, 
    p.MSRP
FROM 
    products p
    INNER JOIN 
    orderdetails d USING(productCode)
WHERE 
    PriceEach < (MSRP * 0.8)
""")
```

```python
# 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)
run("""
SELECT 
    p.productName,
    p.buyPrice,
    d.priceEach
FROM  
    products p
    INNER JOIN 
    orderdetails d ON p.productCode = d.productCode
WHERE 
    d.priceEach>=2*p.buyPrice

""")
```

```python
# 7. List the products ordered on a Monday.
run("""
SELECT 
    p.productName, 
    o.orderDate, 
    DAYNAME(o.orderDate) as day_name 
FROM 
    orderdetails d
    INNER JOIN
    products p ON d.productCode = p.productCode 
    INNER JOIN
    orders o ON o.orderNumber = d.orderNumber
WHERE 
    dayname(o.orderDate) = 'Monday'
""")
```

```python
# 8. What is the quantity on hand for products listed on 'On Hold' orders?
run("""
SELECT 
    o.orderNumber,
    p.productName,
    o.status,
    d.quantityOrdered
FROM 
    orderdetails d
    INNER JOIN
    products p ON d.productCode = p.productCode
    INNER JOIN
    orders o ON o.orderNumber = d.orderNumber
WHERE 
     o.status = 'On Hold'
""")
```
